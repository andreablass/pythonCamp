```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
dias = ['M', 'T', 'W', 'TH', 'F', 'ST', 'SD']
temp = {'México':[28.5, 30.5, 31, 30, 28, 27.5, 30.5], 'BC':[21.2, 22.5, 23.5, 24, 265.5, 246.5, 28]}
```

```python
plt.figure(figsize=(13,4))

plt.subplot(1,2,2) 
plt.plot(dias, temp['México'], 'r*-', color="purple")
plt.xlabel('Días de la semana')
plt.ylabel('Temperaturas - BC')
plt.grid(True)
plt.tight_layout() 

```

```python
plt.figure(figsize=(13,4))

plt.subplot(1,2,2) 
plt.plot(dias, temp['México'], 'r*-', color="purple")
plt.xlabel('Días de la semana')
plt.ylabel('Temperaturas - BC')
plt.title('Temperaturas diarias')
plt.grid(True)
plt.tight_layout() 
```

```python
plt.figure(figsize=(11,4))

plt.subplot(1,2,2) 
plt.plot(dias, temp['México'], color="purple", lw=3, ls='--', marker='o')
plt.xlabel('Días de la semana')
plt.ylabel('Temperaturas - BC')
plt.title('Temperaturas diarias')
plt.grid(True)
plt.tight_layout()  
```

```python
plt.figure(figsize=(11,4))

plt.subplot(1,2,2) 
plt.plot(dias, temp['México'], color="purple", lw=3, ls=':', marker='s')
plt.xlabel('Días de la semana')
plt.ylabel('Temperaturas - BC')
plt.title('Temperaturas diarias')
plt.grid(True)
plt.tight_layout()
```
