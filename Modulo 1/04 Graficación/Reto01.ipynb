Replica la gráfica de línea utilizando el color azul y el marcador de circulo, todo esto dentro de una gráfica con cuadricula

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
x=[1,3,5,6]
y= [1,3,5,1.7]
```

```python
fig, ax=plt.subplots()
plt.xlabel('Eje x')
plt.ylabel('Eje y')
plt.title('Gráfica de línea')
ax.plot(x,y, 'b-', marker='o', markersize=10, markerfacecolor='pink')
plt.grid(True)
plt.show()
```
