## **4.1 Características de la visualización de datos**

La visualización de datos con **Python** es una habilidad necesaria en la ciencia de datos. En general, consiste en representar de forma gráfica los datos para interpretarlos de manera visual, con el fin de comunicar los datos, especialmente cuando se tiene un gran conjunto de estos.

El objetivo de visualizar los datos es captar el interés y mantenerlo por medio de visualizaciones. Cuando vemos un gráfico, vemos rápidamente las distribuciones, las tendencias y los valores fuera de lo normal. De esa forma, podemos ver algo de manera inmediata.

Existe una gran variedad de visualizaciones que se pueden utilizar, pero abordaremos el tema de qué gráfica representa el mensaje que queremos transmitir; por ejemplo, cuando queremos mostrar la volumetría de los datos, lo más adecuado sería utilizar alguna gráfica de pastel, a fin de contrastar la rebanada con el total del círculo.

A continuación se presentan las principales librerías para visualización de datos en **Python**.

 - **Matplotlib**: es la librería gráfica de **Python** estándar y la más conocida. Puedes usar matplotlib para generar gráficos de la calidad necesaria para publicarlos tanto en papel como digitalmente.

Podemos generár un *subplot* de diferentes tamaños, lo cual se puede lograr con el metodo .subplot2grid().

```python
plt.subplot2grid((3, 3), (0, 0), colspan=3, rowspan=1)
plt.subplot2grid((3,3), (1,0), colspan=2)
plt.subplot2grid((3,3), (1,2), rowspan=2)
plt.subplot2grid((3,3), (2,0))
plt.subplot2grid((3,3), (2,1))
plt.tight_layout() # para que no se superpongan las graficas
```

Se pueden configurar los rangos de los ejes con los métodos **ylim** y **xlim** en el objeto del eje, o con **axis('tight')** para obtener automáticamente rangos de ejes **“tightly fitted”**:

```python
plt.subplot2grid((3,3), (0,0), colspan=3)
plt.plot(x, np.sin(x), 'r--')
plt.title('Seno')

plt.subplot2grid((3,3), (1,0), colspan=2)
plt.plot(x, np.cos(x), 'g*-')
plt.title('Coseno')

plt.subplot2grid((3,3), (1,2), rowspan=2)
plt.plot(x, np.exp(x), 'b.-')
plt.title('Exponencial')

plt.subplot2grid((3,3), (2,0))
plt.plot(x, np.log(x), 'k--')
plt.title('Logarítmica')

plt.subplot2grid((3,3), (2,1))
plt.plot(x, x, 'y--')
plt.title('Recta')

plt.tight_layout() # para que no se superpongan las graficas

```

## **4.2 Gráficos de dispersión**

Los gráficos de dispersión se utilizan para comprobar la relación entre dos variables; generalmente se utiliza este tipo de gráficos para estudiar las relaciones que existen entre dos factores, problemas o causas mediante las variables independientes y dependientes.

## **4.3 Gráficos de línea**

Los gráficos de líneas se utilizan para visualizar los cambios a lo largo de un rango continuo, como el tiempo o la distancia. La visualización del cambio con un gráfico de líneas permite ver la tendencia general y, al mismo tiempo, comparar simultáneamente varias tendencias, es decir, con dos o más líneas a la vez, para ver las tendencias y las similitudes que tienen.

```python
fig, ax = plt.subplots()
ax.plot([1, 2, 3, 4], [1, 2, 0, 0.5])
plt.show()
```

## **4.4 Gráficos de área**

Los gráficos de área nos permiten identificar tendencias en los datos en el transcurso del tiempo o en algún rango continuo. Son parecidos a los gráficos de líneas; se diferencian en que el espacio situado debajo de las líneas está sombreado, para que se aprecie mejor la magnitud de las tendencias, con lo cual se logra resaltar el área que esta tendencia representa.

```python
fig, ax = plt.subplots()
ax.fill_between([1, 2, 3, 4], [1, 2, 0, 0.5])
plt.show()
```

## **4.5 Gráfico de barras**

Un gráfico de barras es una forma de resumir un conjunto de datos mediante categorías. Muestra los datos usando varias barras de la misma anchura, cada una de las cuales representa una categoría concreta. La altura de cada barra es proporcional a una agregación específica, por ejemplo, la suma de los valores de esas categorías.

```python
fig, ax = plt.subplots()
ax.bar([1, 2, 3], [3, 2, 1])
plt.show()
```

## **4.6 Histogramas**

Los histogramas son gráficos que indican la frecuencia de un hecho mediante una distribución de los datos. Esto ayuda a observar la distribución de alguna variable, por ejemplo, si alguna variable tiene alguna distribución conocida, como la normal.

```python
fig, ax = plt.subplots()
x = np.random.normal(5, 1.5, size=1000)
ax.hist(x, np.arange(0, 11))
plt.show()
```

## **4.7 Gráficos de pastel**

Una gráfica de pastel es circular y se divide en rebanadas. Ilustra magnitudes o frecuencias relativas. El área de cada sector es proporcional a la cantidad que representa y, en conjunto, las rebanadas crean un círculo completo.

```python
fig, ax = plt.subplots()
ax.pie([5, 4, 3, 2, 1])
plt.show()
```

## **4.8 Gráficos de caja y bigote**

Un diagrama de cajas y bigotes es una manera conveniente de mostrar visualmente grupos de datos numéricos por medio de sus cuartiles.

Las líneas que se extienden paralelas a las cajas se conocen como *bigotes*, y se usan para indicar la variabilidad fuera de los cuartiles superior e inferior.

```python
fig, ax = plt.subplots()
ax.boxplot([1, 2, 1, 2, 3, 4, 3, 3, 5, 7])
plt.show()
```

## **4.9 Gráficos de violín**

Los gráficos de violín se utilizan para mostrar la distribución y la densidad de la probabilidad de los datos; es una combinación de un diagrama de caja y un gráfico de densidad. La parte central del gráfico de violín refleja la información del diagrama de caja, y los dos lados del gráfico reflejan la información del mapa de densidad. Se utiliza a menudo, por ejemplo, en los análisis exploratorios de datos (EDA) antes de modelar.

```python
fig, ax = plt.subplots()
ax.violinplot([1, 2, 1, 2, 3, 4, 3, 3, 5, 7])
plt.show()
```

## **4.10 Mapas de color**

Un mapa de color es un conjunto de valores que están asociados con colores; se utilizan para mostrar un ráster de banda única de manera coherente con los mismos colores. Por ejemplo, cuando quieres detectar las correlaciones entre las variables de un *set* de datos.

```python
fig, ax = plt.subplots()
x = np.random.random((16, 16))
ax.imshow(x)
plt.show()
```

### Matplotlib

Esta libreria inicio con su primera versión en el año 2003 y ha tenido 4 versiones distintas a la largo de su desarrollo, llegando a Matplotlib 3 en su último lanzamiento.

A lo largo de la trayectoría y en el uso diario de la libreria "Matplotlib", encontraremos 2 distintas aproximaciones dependiendo su versión, en un inicio cuando se creo con su propietario MATLAB esta estaba orientada hacia una programación orientada a estados (POS),más que a una programación orientada a objetos (POO), esto implicaba algunas desventajas, por ejemplo, era díficil encapsular objetos y funciones, como resultado colocando más código para funciones más complejas.

La idea inicial de Matplotlib fue hacer un puente entre MATLAB y Python, con el lanzamiento de las versiones fue evolucionando hacia una POO, para ejemplicar lo dicho pondremos un ejemplo con una POO y un POS.

Aquí volvemos a poner los valores iniciales a X para no tener errores de aquí en adelante.

Python es un lenguaje de tipado dinamico, es decir, que puedes cambiar el tipo del dato de una variable, y su valor, a lo largo de todo el programa.

```python
x = np.linspace(0,5, 11)
```

```python
#POS
plt.figure(facecolor='white')
plt.subplot(2,2,1)
plt.plot(x, y, 'r-')
plt.subplot(2,2,2)
plt.plot(x, y, 'b-')
plt.subplot(2,2,3)
plt.plot(x, y, 'g-')
```

Creamos una figura con 3 gráficas dentro, observemos que la sentencia "plt.subplot(2,2,1)" crea y se enfoca en el primer grafico, yendo de izquierda a defrecha en una cuadrícula 2x2.

Después la sentencia "plt.subplot(2,2,1)" grafica el siguiente gráfico en la izquierda tomando como principal el primer gráfico.

```python
#POO
fig, ax = plt.subplots(2,2)
fig.set_facecolor('white')
ax[0,0].plot(x, y, 'r-')
ax[0,1].plot(x, y, 'b-')
ax[1,0].plot(x, y, 'g-')
```

Para este caso, se declaro una figura con la sentencia "fig" y "ax" de una cuadrícula 2x2, la diferencia rádica que para las gráficas, iniciamos declarando por medio de "ax" la posición en donde queremos que se coloque la gráfica.

Ambas formas llegarán al mismo resultado, pero es importante conocer ambas formas para entender porque se puede gráficar de una u otra forma.
